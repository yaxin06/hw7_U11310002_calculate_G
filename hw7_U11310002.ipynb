```python
import math

# Constants
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
radius_earth = 6371000  # Radius of Earth in meters
density_bouguer = 2670  # Density for Bouguer correction in kg/m^3

# Function to calculate the theoretical gravity (ignoring the correction)
def theoretical_gravity(latitude, elevation):
    # Formula for theoretical gravity in m/s^2
    # G = gravitational constant, 
    # r = radius of Earth, 
    # latitude = latitude of the location
    # elevation = elevation above sea level in meters
    lat_rad = math.radians(latitude)  # Convert latitude to radians
    g_theory = 9.780327 * (1 + 0.00193185138639 * math.sin(lat_rad) ** 2) / math.sqrt(1 - 0.00669437999014 * math.sin(lat_rad) ** 2)
    g_theory = g_theory * (1 + (elevation / radius_earth))  # Correction for elevation
    return g_theory

# Function to calculate free air correction (in mGal)
def free_air_correction(elevation):
    # Free-air correction in mGal, 1 mGal = 10^-5 m/s^2
    return 0.3086 * elevation  # Free-air correction formula

# Function to calculate Bouguer correction (in mGal)
def bouguer_correction(elevation):
    # Bouguer correction in mGal, 1 mGal = 10^-5 m/s^2
    return 0.0419 * elevation * density_bouguer  # Bouguer correction formula

# Function to calculate gravity anomaly (in mGal)
def gravity_anomaly(observed_gravity, theoretical_gravity, free_air_corr, bouguer_corr):
    return observed_gravity - (theoretical_gravity + free_air_corr + bouguer_corr)

# Main code to calculate all components
def main():
    # Given data
    latitude = 48.1195  # Latitude in degrees
    longitude = 12.1878  # Longitude in degrees
    elevation = 487.9  # Elevation in meters
    observed_gravity = 980717.39  # Observed gravity in mGal
    
    # Calculations
    theo_gravity = theoretical_gravity(latitude, elevation)
    free_air_corr = free_air_correction(elevation)
    bouguer_corr = bouguer_correction(elevation)
    
    # Gravity anomalies
    bouguer_gravity_anomaly = gravity_anomaly(observed_gravity, theo_gravity, free_air_corr, bouguer_corr)
    
    # Print results
    print(f"Theoretical Gravity: {theo_gravity:.6f} m/s^2")
    print(f"Free Air Correction: {free_air_corr:.2f} mGal")
    print(f"Bouguer Correction: {bouguer_corr:.2f} mGal")
    print(f"Gravity Anomaly (Bouguer): {bouguer_gravity_anomaly:.2f} mGal")
    
    # (b) Scenario assuming the gravity station is over the ocean
    water_depth = 487.9  # Water depth in meters
    free_air_corr_ocean = free_air_correction(water_depth)
    bouguer_corr_ocean = bouguer_correction(water_depth)
    
    # Gravity anomalies for ocean scenario
    bouguer_gravity_anomaly_ocean = gravity_anomaly(observed_gravity, theo_gravity, free_air_corr_ocean, bouguer_corr_ocean)
    
    # Print ocean results
    print(f"Free Air Correction (Ocean): {free_air_corr_ocean:.2f} mGal")
    print(f"Bouguer Correction (Ocean): {bouguer_corr_ocean:.2f} mGal")
    print(f"Gravity Anomaly (Bouguer, Ocean): {bouguer_gravity_anomaly_ocean:.2f} mGal")

# Run the main function
if __name__ == "__main__":
    main()
```
